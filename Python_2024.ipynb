# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Criação de figura com dados
fig, ax = plt.subplots()
ax.plot([1,10,100,1000], [1,2,3,4])
plt.show()
```

```python
#Figura com grid de eixos
fig, axs = plt.subplots(2,2)
plt.show()
```

```python
# Outras configurações de subplots
fig, axs = plt.subplot_mosaic([['left_top', 'right'], ['left_bottom', 'right']])
plt.show()
```

```python
#Escalas dos eixos
fig, axs = plt.subplot_mosaic([['linear', 'linear-log'], ['log-linear', 'log-log']], layout='constrained')

x = np.arange(0, 4*np.pi, 0.1)
y = 2 * np.sin(x) + 3

#Para o primeiro subplot
ax = axs['linear']
ax.plot(x, y)
ax.set_xlabel('Linear')
ax.set_ylabel('Linear')
fig.show()

#Para o segundo subplot:
ax = axs['linear-log']
ax.plot(x, y)
ax.set_yscale('log')
ax.set_xlabel('Linear')
ax.set_ylabel('Log')
fig.show()

#Para o terceiro subplot:
ax = axs['log-linear']
ax.plot(x,y)
ax.set_xscale('log')
ax.set_xlabel('Log')
ax.set_ylabel('Linear')
fig.show()

#Para o quarto subplot:
ax=axs['log-log']
ax.plot(x,y)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Log')
ax.set_ylabel('Log')
fig.show()
```

```python
# Argumentos para escalas (figsize):

fig, axs = plt.subplot_mosaic([['log', 'symlog']], layout='constrained', figsize=(5, 4))

for name, ax in axs.items():
    if name in ['log']:
        ax.plot(x, x)
        ax.set_yscale('log',base=2)
        ax.set_title('Log base=2')
        fig.show()
    else:
        ax.plot(x - np.mean(x), x - np.mean(x))
        ax.set_yscale('symlog',)
        ax.set_title('Symlog')
        fig.show()
        
```

```python
#Grid e limite dos eixos:

fig, axs = plt.subplot_mosaic([['log', 'symlog']], layout='constrained', figsize=(5, 4))

for name, ax in axs.items():
    if name in ['log']:
        ax.plot(x, x)
        ax.set_yscale('log',base=2)
        ax.set_title('Log base=2')
        ax.grid(True) #Acionar o grid
        ax.set_xlim([0,15]) #Definir o limite do eixo x
        ax.set_ylim([0,16]) #Definir o limite do eixo y
        
    else:
        ax.plot(x - np.mean(x), x - np.mean(x))
        ax.set_yscale('symlog',)
        ax.set_title('Symlog')
        print('O eixo x é:', ax.xaxis.get_scale()) #Para saber qual é o eixo
        print('O eixo y é:', ax.yaxis.get_scale()) #Para saber qual é o eixo
```

```python
np.random.seed(19680801)

data = {'a': np.arange(50), 'c': np.random.randint(0,100,50) }
data['b'] = np.abs(data['a']+ 10*np.random.randn(50))
data['d'] = np.abs(np.random.randn(50))*10


#print('Os valores da chave d são: ','\n', data['d'] )
#Números aleatórios inteiros com valores entre 0-99 e 50 valores.
#print(data['a']) #Lembrete de como printar os valores

fig, axs = plt.subplot_mosaic([['Normal','Dispersão']],figsize=(6,4), layout='constrained')

#Primeiro gráfico: Normal
ax = axs['Normal']
ax.plot(data['a'],data['b'])
ax.set_title('Normal')
ax.set_xlabel('Entry x')
ax.set_ylabel('Entry y')
fig.show()

#Segundo gráfico: Dispersão
ax = axs['Dispersão']
ax.scatter('a', 'b', data=data, c='c', s='d') #Nesse exemplo o parâmetro "c" cuida das cores, definindo um padrão de cores. O parâmetro "d" cuida do tamanho dos pontos
fig.show()
```

```python
x = np.linspace(0,2,100)

fig, ax = plt.subplots(figsize=(5,4), layout='constrained')

ax.plot(x, x, label='linear', color= 'blue') #adcionar uma label para a linha
ax.plot(x, x**2, label='quadratic', color='orange')
ax.plot(x, x**3, label='cubic', color ='green')

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Exemplo')

ax.legend() #Colocar legendas
plt.show()
```

```python
# Quadro de resumo de funcionalidades
fig, axs = plt.subplots(ncols=2, nrows=1, layout='constrained')

#Primeira Figura:
ax= axs[0] #Para plotar na primeira figure
ax.set_title('Exemplo 1')
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')
x = np.linspace(-2,2,1000)
ax.plot(x, x**2, label=r'$x^2$', color='red', linewidth=1.5,alpha=1 ) #Exemplo de parâmetros
ax.legend(loc='upper center')
#fig.savefig('Inserir nome', dpi=400) tem que vim antes do fig.show()
fig.show()

#Segunda Figura:
ax= axs[1]
x = np.random.randint(0,100,50)
y = np.random.randint(0,200,50)
ax.set_title('Exemplo 2')
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')
ax.scatter(x,y, s=2.5)
ax.axhline(100, linestyle='dashed', color='orange')
ax.axvline(25, linestyle='dotted', color='pink')
fig.show()
```

# Progamação orientada a objetos

Classe: Representa uma série de funções com determinadas funcionalidades
Atributo: São as variáveis das funções que estão dentro das funções
Método: São definidos pelas funções e podem ser chamados

```python
class OperacoesNumeros():
    def incremento(self, n,i):
        valor_i = n 
        valor_f = n + i
        return valor_f
```

```python
obj = OperacoesNumeros() #A classe foi instânciada

## Agora pode-se usar os métodos da classe para cada objeto

print(obj.incremento(10,1)) #1 maneira
obj1 = obj.incremento(10,1)
print('Maneira atribuindo a uma variável', obj1)

print(OperacoesNumeros().incremento(12,2)) #2 maneira
obj2 = OperacoesNumeros().incremento(12,2)
print('Maneira atribuindo a uma variável', obj2)
```
